```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('/home/sairam/projects/evv2gfcr/evv2gfcr/FrequencyGenerator/FrequencyData/germany_2020_01.csv.zip', index_col=0)

print(df)
```

```python
df['Frequency'] = df['Frequency']/1000 + 50
average = df['Frequency'].mean()
df['Frequency'].fillna(average, inplace=True)
print(df)
```

```python
# Plotting
plt.figure(figsize=(10,6))
plt.plot(df['Frequency'].iloc[:1000])
plt.title('Frequency Plot')
plt.xlabel('Index')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
